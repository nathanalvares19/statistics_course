```python
import numpy as np
import pandas as pd

population = pd.read_excel("population_data.xlsx")
sample = pd.read_excel("sample_data.xlsx")

pop_salary = population["Salary"].to_numpy()
sample_salary = sample["Annual Salary($)"].to_numpy()
```

```python
pop_mean = np.mean(pop_salary)
pop_var = np.var(pop_salary)

print(pop_mean, pop_var)
```

```python
sample_mean = np.mean(sample_salary)

sample_var_biased = np.var(sample_salary, ddof=0)   
sample_var_unbiased = np.var(sample_salary, ddof=1) 

print(sample_mean, sample_var_unbiased)
```
