```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Read excel data
population = pd.read_excel("population_data.xlsx")
sample = pd.read_excel("sample_data.xlsx")

# Plot histograms
population["Salary"].plot(kind="hist", bins=25, alpha=1, label="Population")
sample["Annual Salary($)"].plot(kind="hist", bins=25, alpha=1, label="Sample")
plt.xlabel("Salary")
plt.ylabel("Frequency")
plt.legend()
plt.title("Population vs Sample Salary Histogram")
plt.show()

sample["Annual Salary($)"].plot(kind="hist", bins=25, alpha=1, label="Sample")
plt.xlabel("Salary")
plt.ylabel("Frequency")
plt.legend()
plt.title("Sample Salary Histogram")
plt.show()

pop_salary = population["Salary"].to_numpy()
sample_salary = sample["Annual Salary($)"].to_numpy()
```

```python

# Population parameters
pop_mean = np.mean(pop_salary)
pop_var = np.var(pop_salary)
pop_stddev = np.std(pop_salary) 

print(f"Population mean: {pop_mean}\nPopulation variance: {pop_var:.02f}\nPopulation standard deviation: {pop_stddev:.02f}")
```

```python

# Sample statistics
sample_mean = np.mean(sample_salary)
sample_var_biased = np.var(sample_salary, ddof=0) # 1 / n   
sample_var_unbiased = np.var(sample_salary, ddof=1) # 1 / (n - 1)
sample_stddev_biased = np.std(sample_salary, ddof=0) # 1 / n
sample_stddev_unbiased = np.std(sample_salary, ddof=1) # 1 / (n - 1)

print(f"Sample mean: {sample_mean}\nSample variance (unbiased): {sample_var_unbiased:.02f}\nSample variance (biased): {sample_var_biased:.02f}\nSample standard deviation (unbiased): {sample_stddev_unbiased:.02f}\nSample standard deviation (biased): {sample_stddev_biased:.02f}")
```

```python
# Generate random samples
n = 30 # sample size
N = 10_000 # number of repetitions

sample_means = []
sample_sds = []

for _ in range(N):
    sample = np.random.normal(pop_mean, pop_stddev, n)
    sample_means.append(np.mean(sample))
    sample_sds.append(np.std(sample, ddof=1))

sample_means = np.array(sample_means)
sample_sds = np.array(sample_sds)

# Plot histogram of sample means
plt.hist(sample_means, bins=40, density=True)
plt.axvline(pop_mean, linewidth=2, color="red")  # true mean
plt.xlabel("Sample Mean")
plt.ylabel("Density")
plt.title("Sampling Distribution of the Mean")
plt.show()

# Plot histogram of sample standard deviations
plt.hist(sample_sds, bins=40, density=True)
plt.axvline(pop_stddev, linewidth=2, color="red")  # true standard deviation
plt.xlabel("Sample Standard Deviation")
plt.ylabel("Density")
plt.title("Sampling Distribution of the Standard Deviation")
plt.show()
```
